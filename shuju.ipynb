```python
def marriage(y,list):
    for i in range(1,y+1):
        if i==1:
            list.append(max((y-i)/(y+1-i)*0+1/y,0))
        else:
            list.append(max((y-i)/(y+1-i)*list[i-2]+1/y,list[i-2]))
    
    return list
```

```python
N = 10
E = [[0] * (N + 1) for _ in range(N + 1)]

# Base Case
for j in range(N + 1):
    E[N][j] = 1

# Recursive Case
for i in range(N - 1, -1, -1):
    for j in range(i, -1, -1):
        E[i][j] = max(E[i + 1][j + 1], E[i + 1][j])

# Optimal expected wealth
optimal_wealth = E[0][0]
print("Optimal Expected Wealth:", optimal_wealth)
```
